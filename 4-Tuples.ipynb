Tuple items are ordered, unchangeable, and allow duplicate values. When we say that tuples are ordered, it means that the items have a defined order, and that order will not change. Tuples are unchangeable, meaning that we cannot change, add or remove items after the tuple has been created. It allow Duplicates, Tuple items can be of any data type. Represented by () but have some exception.

```python
# creat an empty tuple
tup = ()
tup
type(tup)
```

```python
sample = (1, 2, 3.5, "apple", True,1,3.5)
print(sample)
type(sample)
```

```python
# access the elements of tuple
print(sample[2])
print(sample[4])
print(sample[-1])
```

```python
lis = [1, 2, 3]
lis[1] = 5 # assignment is possible in list but not in tuple
lis
```

```python
# negative slicing
sample = (1, 2, 3.5, "apple")
print(sample[-2])
print(sample[-3:])  
print(sample[ :-1])
```

```python
#Slicing of tuple
```

```python
tup = (1,2,3,4)
print(tup[::-1]) # reverse
```

```python
t = (9,3,2,5,7,3,7)
print(t[2:7])
```

```python
t = (1, 2, 3, 4, 5, 6)
print(t[::2])
```

Once a tuple is created, you cannot change its values. Tuples are unchangeable, or immutable as it also is called.
But there is a workaround. You can convert the tuple into a list, change the list, and convert the list back into a tuple.

```python
#typecast tuple to the list and back to tuple
x = ("apple","pineapple", "cherry")
y = list(x)
```

```python
print(y)
print(type(y))
```

```python
t = ('a', 'b', 'c')
temp = list(t)
temp.append('d')  # Add element
t = tuple(temp)
print(t)  
```

```python
#use of zip() and list of tuples
names = ['Urvashi', 'Prasanna', 'Gayatri']

age = [23, 78, 12]

zip(names, age)
```

```python
info = list(zip(names, age))
info
```
