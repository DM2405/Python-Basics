Conditional Statements
Conditional Statements are also known as decision making statements.We need to use these conditional statements to execute the specific block of code if the given condition is true or false.

In Python we can achieve this by using

if statements
if - else statements
elif statements
Nested if and if-else statements

```python
text = "hello"
if "a" in text:

  print("a is present in the text")

else:
  print(" a is not present in the text")
```

```python
x = int(input("Enter a number: "))  # user input

if x % 2 == 0:
  print("x is an even number")
else:
  print("x is an odd number")
```

```python
x = 9
y = 2
if x < y:
  print("x is less than y")
elif x > y:
  print(" x is greater than y")
else:
  print("True")
```

```python
x = int(input("Enter a number"))

if x > 0:
  print("x is greater than zero")
elif x < 0:
  print("x is less than zero")
else:
  print("Zero")
```
